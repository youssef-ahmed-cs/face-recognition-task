### ------------------------------------------------------------
### Data Augmentation
### ------------------------------------------------------------
- Data Augmentation is a technique used in computer vision to artificially increase the size and diversity of a training dataset by        generating modified versions of existing images.

- Instead of collecting new images, we apply small transformations to the original images so the model can see different variations
    of the same data during training.

- The main goal is to help the model generalize better and avoid overfitting (when the model memorizes the training data instead
     of learning general patterns).

In real-world scenarios, images of the same object can appear
under different conditions such as:
- different lighting
- slight rotations
- camera distance
- mirrored views

### Data augmentation simulates these variations so the model becomes more robust.

### Common augmentation techniques include:

rotation_range
  - randomly rotates the image within a small angle range

zoom_range
 - randomly zooms in or out of the image

brightness_range
  - randomly changes the image brightness to simulate
  different lighting conditions

horizontal_flip
  - flips the image horizontally to simulate mirrored views



```python
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
import os
```

```python
datagen = ImageDataGenerator(
    rotation_range=10,
    zoom_range=0.1,
    brightness_range=[0.8,1.2],
    horizontal_flip=True
)
```

```python
import shutil
def augment_images(input_dir, output_dir, num_augmented_images):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    for filename in os.listdir(input_dir):

        img_path = os.path.join(input_dir, filename)

        img = load_img(img_path)

        x = img_to_array(img)
        
        x = x.reshape((1,) + x.shape)  #1*224*244*3

        i = 0
        for batch in datagen.flow(x, batch_size=32, save_to_dir=output_dir, save_prefix='aug', save_format='jpg'):
            i += 1
            if i >= num_augmented_images:
                break  
```

```python
base_dir = r"D:\4- Simple Face Recognition\persons"
output_root = "Face Recoginition/augmented_data"
num_augmented_images = 5

for class_dir in os.listdir(base_dir):

    input_dir = os.path.join(base_dir, class_dir)
    class_output_dir = os.path.join(output_root, class_dir)

    augment_images(input_dir, class_output_dir, num_augmented_images)
```
